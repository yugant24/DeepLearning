### Many of columns has a categorical values lets Check it

```python
def print_unique_columns(df):
    for column in df:
        if df[column].dtype == 'object':
             print(f'{column}: {df[column].unique()}') 
```

```python
print(print_unique_columns(df1))
```

```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
print_unique_columns(df1)
```

### convert Yes No to 1 and 0

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df1[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
df1.dtypes
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace =True)
```

```python
df1.gender.unique()
```

```python
df2 = pd.get_dummies(data=df1,columns=['InternetService','Contract','PaymentMethod']).astype('int')
df2.sample(3)
```

```python
df2.dtypes
```

```python
df2.head(2)
```

```python
cols_toscale=['tenure','MonthlyCharges','TotalCharges']
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df2[cols_toscale]=scaler.fit_transform(df2[cols_toscale])
```

```python
for col in df2:
    print(f'{col}: {df2[col].unique()}')
```

### now lets train a model

```python
X=df2.drop('Churn',axis=1)
y=df2['Churn']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
X_train.shape
```

```python
len(X_train.columns)
```

### Build an ANN network


```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(26,input_shape=(26,),activation = 'relu'),
    keras.layers.Dense(20,activation = 'relu'),
   
    keras.layers.Dense(19,activation = 'relu'),
    keras.layers.Dense(15,activation = 'relu'),
    keras.layers.Dense(1,activation = 'sigmoid')
])
model.compile(optimizer='adam',
              loss = 'binary_crossentropy',
              metrics=['accuracy'])
model.fit(X_train, y_train, epochs=200)
```

```python
model.evaluate(X_test,y_test)
```

```python
yhat = model.predict(X_test)
```

```python
yhat[:6]
```

```python
y_test[:6]
```

```python
y_pred=[]
for ele in yhat:
    if ele>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
np.array(y_test.head(10))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
      
```

```python
import seaborn as sns
cm =tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
